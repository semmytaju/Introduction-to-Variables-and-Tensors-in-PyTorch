# **GPU Acceleration**

```python
torch.cuda.is_available()
```

```python
# Select Device
device = torch.device(
    "cuda" if torch.cuda.is_available()
    else "cpu"
)

print(device)
```

```python
# Move Tensor to GPU
x = torch.rand((3,3))

x = x.to(device)

print(x.device)
```

```python
# Mini Project
# Implement a simple linear model: y=wx+b

w = torch.tensor(
    2.0,
    requires_grad=True
)

b = torch.tensor(
    1.0,
    requires_grad=True
)

x = torch.tensor(3.0)

y = w*x + b

print(y)
```
