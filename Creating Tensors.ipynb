**Creating Tensors**

```python
# Scalar Tensor
import torch

x = torch.tensor(10)

print(x)
```

```python
# Vector Tensor
x = torch.tensor([1,2,3])

print(x)
```

```python
# Matrix Tensor
x = torch.tensor([
    [1,2],
    [3,4]
])

print(x)
```

```python
# Tensor Shape (Shape describes the size of each dimension.)
x = torch.tensor([
    [1,2,3],
    [4,5,6]
])

print(x.shape)

# 2 rows
# 3 columns
```

```python
# Tensor Dimensions
x = torch.tensor([
    [1,2,3],
    [4,5,6]
])

print(x.ndim)
```

```python
# Tensor Data Types (Every tensor stores values using a specific data type.)
x = torch.tensor([1,2,3])

print(x.dtype)
```

Common dtypes:
# Type	      Description
*   torch.int32	= Integer
*   torch.int64	= Long Integer
*   torch.float32	= Float
*   torch.float64	= Double
*   torch.bool	= Boolean

# **Creating Tensors with Different Initializations**

```python
# Zeros
torch.zeros((3,3))
```

```python
# Ones
torch.ones((3,3))
```

```python
# Random Uniform
torch.rand((3,3))
```

```python
# Random Gaussian
torch.randn((3,3))
```
