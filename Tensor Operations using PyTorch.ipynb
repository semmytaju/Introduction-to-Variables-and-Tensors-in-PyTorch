**Matrix Multiplication**
Deep learning heavily relies on matrix multiplication.

```python
A = torch.tensor([
    [1,2],
    [3,4]
], dtype=torch.float32)

B = torch.tensor([
    [5,6],
    [7,8]
], dtype=torch.float32)

C = torch.matmul(A,B)

print(C)
```

**Tensor Reshaping**

**Automatic Differentiation (Autograd)**
One of PyTorch's most powerful features.
Autograd automatically computes derivatives required during neural network training.

Consider:

y=x^2

```python
# Creating a Differentiable Tensor
x = torch.tensor(
    2.0,
    requires_grad=True
)
```

```python
# Computing Gradients
y = x**2

y.backward()

print(x.grad)
```
