**What is a Tensor?**
A Tensor is the fundamental data structure in PyTorch. Mirip dengan: Scalar (0D), Vector (1D), Matrix (2D) dan Tensor or Multi-dimensional Array (nD). Visualisasi:

Scalar     -> 5

Vector     -> [1,2,3]

Matrix     -> [[1,2],
               [3,4]]

Tensor 3D  -> [[[...]]]

**Why Tensors Matter in Deep Learning? **Every component of a neural network is represented as tensors:
*   Inputs
*   Weights
*   Biases
*   Activations
*   Outputs
*   Gradients

```python
# Scalar
x = torch.tensor(10)

print(x)
```

```python
# Vector
x = torch.tensor([1,2,3])

print(x)
```

```python
# Matrix
x = torch.tensor([
    [1,2],
    [3,4]
])

print(x)
```
